# Question 6
-Under Inaugural Address package
-For every president, find out how many unique words they used in speech. 
-We will just assume that President Bush means one person (technically it is incorrect but we will ignore it)


```python
import nltk
from nltk.corpus import inaugural

files = inaugural.fileids()
#print('files \n',files)
presedential_list = []

for file in files:
    new_name = file[5:-4]
    if new_name not in presedential_list:
        presedential_list.append(new_name)
    
#print(presedential_list)
```

```python
president_map = dict()

for president in presedential_list:
    for file in files:
        if president in file:
            if president in president_map:
                unique_words = president_map[president]
            else:
                unique_words = set()
            
            word_set = inaugural.words(file)
            for word in word_set:
                if word.isalpha() and word.lower() not in unique_words:
                    unique_words.add(word.lower())
                #elif not word.isalpha() and word not in unique_words:
                    #unique_words.add(word)
            president_map[president] = unique_words

print('President unique words(removed digits and special characters) \n')    
    
for k in president_map:
    print('\n',k,' : count ',len(president_map[k]),'\n')
    print(president_map[k])
    
```
